```python
from ibm_watson_machine_learning import APIClient
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# assets = wslib.assets.list_assets("asset")
# print(assets)

# Accessing asset
assets_by_name = wslib.assets.list_assets("data_asset", name="dataset_sample.csv")
print(assets_by_name)

my_file = wslib.load_data('dataset_sample.csv',attachment_type_or_item=None )

```

```python
my_file.seek(0)
import pandas as pd 
data = pd.read_csv(my_file)
```

```python
# split the data
# x = features
# y = target
x = data[['Fan','Refrigerator','AirConditioner','Television','Monitor','MotorPump']]
y = data["Weekly CO2 Emission (Kgs)"]

# Split data 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate the mean squared error 
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

future_data = pd.DataFrame({'Fan': [12], 'refrigerator': [20], 
                            'airconditioner': [30], 'television': [10], 
                            'monitor': [10], 'motorpunp': [20]})
future_emission = model.predict(future_data)
print(f"Predicted Future Carbon Emission: {future_emission[0]} kgs")



```
